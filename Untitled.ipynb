```python
recList = []
lenList = []

for i in df["path"]:
    #print(i)
    
    sr, rec = read(i)
    recList.append(rec)
    lenList.append(len(rec))
    #print(len(rec))
    #plt.plot(rec)
len(lenList)
```

```python
import numpy as np
plt.xlabel('Seconds')
plt.ylabel('Frequency')
plt.hist(np.array(lenList)/16000, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=1)
```

```python
n = 0
longRecs = []
secList = np.array(lenList)/16000
for i in range(len(secList)):
    if(secList[i] > 25):
        n+=1
        #print(df.loc[i, "action"])
        #print(df.loc[i, "object"])
        #print("-------")
        #print(i)
        longRecs.append(i)

print(longRecs)

```
